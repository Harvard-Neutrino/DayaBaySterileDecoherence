```python
def LogSpacePoints(Emin,Emax,binnum = 30):
	"""Returns log space points.
        
        @type   Emin     :      float
        @param  Emin     :      min value
        @type   Emax     :      float
        @param  Emax     :      max value
        
        @rtype           :      list
        @return          :      log spaced values
	"""
	E_list = []
	Emin_log, Emax_log = np.log(Emin),np.log(Emax)
	if Emin_log == -float('inf') :
		Emin_log = 0.0
	Estep_log = (Emax_log-Emin_log)/float(binnum)
	Elog = np.arange(Emin_log,Emax_log+Estep_log/2.0,Estep_log)
	for logE in Elog:
		E_list.append(np.exp(logE))
    
	return E_list
```

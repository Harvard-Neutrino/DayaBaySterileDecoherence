```python
import sys
import os
common_dir = '/Common_cython'
main_dir = os.getcwd()
#sys.path.append(main_dir+common_dir)

import matplotlib.pyplot as plt
import numpy as np
import matplotlib

path_to_style = main_dir + common_dir
dir = 'PlotData/'
plt.style.use(path_to_style+r"/paper.mplstyle")
matplotlib.rcParams.update({'text.usetex': True})
matplotlib.rcParams['text.latex.preamble']=[r"\usepackage{amsmath}"]
```

```python
size = (11,7)
margins = dict(left=0.14, right=0.96,bottom=0.1, top=0.93)

plot, ax = plt.subplots(figsize = size,gridspec_kw=margins)
ax.grid(linestyle = '--')
ax.tick_params(axis='x')
ax.tick_params(axis='y')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylabel(r"$L (\text{GeV}^{-1})$")
ax.set_xlabel(r"$E (\text{GeV})$")
ax.set_xlim([5e-4,10])
ax.set_ylim([1e16,1e23])

eV = 1e-9
cm = 50677.308/eV
km = 1e5*cm
m  = 1e2*cm
MeV = 1e-3

enu = np.logspace(np.log10(5e-4),np.log10(3e3),500)
Deltam21 = 7.42e-5
Deltam31 = 2.514e-3
Deltam41 = 1

# We are making that the argument is equal to 2pi (so that it matches Jordi's figure)
solar = 4e18*enu/Deltam21*2*np.pi
atm = 4e18*enu/Deltam31*2*np.pi
ste = 4e18*enu/Deltam41*2*np.pi
ax.plot(enu,solar, '--k', enu, atm, '--k',enu,ste,'--k')


sigma = 2.1e-4 #nm
decoh = 5.068e24*4*np.sqrt(2)*enu**2*sigma/Deltam41
ax.plot(enu,decoh)

daya_e = np.logspace(np.log10(1.8e-3),np.log10(1.2e-2),10)
daya_l = np.ones(10)*km
#ax.plot(daya_e,daya_l)
plt.text(2.3e-3,0.58*km,'DayaBay', fontsize = 'small', color = 'navy')
ax.fill_between(daya_e,daya_l*0.3,daya_l*1.5, alpha =0.5)

neos_e = np.logspace(np.log10(1.8e-3),np.log10(1.2e-2),10)
neos_l = np.ones(10)*23.7*m
plt.text(1.81e-3,26*m,'NEOS', fontsize = 'small', color = 'chocolate')
ax.plot(neos_e,neos_l)

mboone_l = np.ones(10)*541*m
mboone_e = np.logspace(np.log10(200*MeV),np.log10(3000*MeV),10)
plt.text(350*MeV,600*m,'MiniBoone', fontsize = 'small', color = 'green')
ax.plot(mboone_e,mboone_l)

daed_l = np.ones(10)*km
daed_e = np.logspace(np.log10(20*MeV),np.log10(40*MeV),10)
plt.text(10*MeV,8*km,r'DAE$\delta$ALUS', fontsize = 'x-small', color = 'sienna')
ax.fill_between(daed_e,daed_l*1.5,daed_l*20, alpha = 0.5)

bugey_l = np.ones(10)*m
bugey_e = np.logspace(np.log10(1.8*MeV),np.log10(12*MeV),10)
plt.text(13*MeV,26*m,'Bugey3', fontsize = 'small', color = 'green')
ax.fill_between(bugey_e,bugey_l*15,bugey_l*40, alpha = 0.5)

danss_l = np.ones(10)*m
danss_e = np.logspace(np.log10(1.8*MeV),np.log10(12*MeV),10)
plt.text(13*MeV,10*m,'DANSS', fontsize = 'small', color = 'purple')
ax.fill_between(bugey_e,bugey_l*10.7,bugey_l*12.7, alpha = 0.5)

kamland_l = np.ones(10)*1e21
kamland_e = np.logspace(np.log10(1.8*MeV),np.log10(8*MeV),10)
plt.text(4e-3,0.8e21,'KamLAND', fontsize = 'small', color = 'darkgoldenrod')
ax.fill_between(kamland_e,kamland_l*0.7,kamland_l*1.2, alpha = 0.5)

plt.show()
```
